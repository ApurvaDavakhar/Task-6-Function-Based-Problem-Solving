Problem Statement 1:Smart Parking Lot Management System

```python
def parking_management(capacity, logs):
    current_vehicles = 0
    peak_usage = 0

    for log in logs:
        if log == "IN":
            current_vehicles += 1
        elif log == "OUT":
            current_vehicles -= 1

        peak_usage = max(peak_usage, current_vehicles)

    if current_vehicles > capacity:
        status = "Parking Full"
    else:
        status = "Available"

    print("Currently Parked Vehicles:", current_vehicles)
    print("Parking Status:", status)
```

```python
parking_management(50, ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"])
```

 Problem Statement 2:Online Food Delivery Time Estimator

```python
def estimate_delivery_time(distance, traffic, weather):
    # Base delivery time based on distance
    time = distance * 5

    # Traffic delay
    if traffic == "High":
        time += 20
    elif traffic == "Medium":
        time += 10

    # Weather delay
    if weather == "Rainy":
        time += 15

    print("Estimated Delivery Time:", time, "minutes")
```

```python
estimate_delivery_time(8, "High", "Rainy")
```

 Problem Statement 3:Movie Theatre Seat Occupancy Analyzer

```python
def analyze_seat_occupancy(total_seats, booked_seats):
    # Count booked seats
    booked_count = len(booked_seats)

    # Calculate occupancy percentage
    occupancy = (booked_count / total_seats) * 100

    # Determine show status
    if occupancy == 100:
        status = "Housefull"
    elif occupancy >= 75:
        status = "Almost Full"
    else:
        status = "Seats Available"

    print("Occupancy:", int(occupancy), "%")
    print("Show Status:", status)
```

```python
booked = [1] * 150   # 150 booked seats
analyze_seat_occupancy(200, booked)
```

Problem Statement 4:Cloud Server Load Classification System

```python
def classify_server_load(cpu_readings):
    total = 0

    # Calculate total CPU usage
    for cpu in cpu_readings:
        total += cpu

    # Calculate average CPU usage
    average_cpu = total / len(cpu_readings)

    # Classify server status
    if average_cpu < 50:
        status = "Normal"
    elif average_cpu <= 80:
        status = "Warning"
    else:
        status = "Critical"

    print("Average CPU Load:", int(average_cpu), "%")
    print("Server Status:", status)
```

```python
cpu_data = [45, 60, 70, 85, 90]
classify_server_load(cpu_data)
```

 Problem Statement 5:Smart Classroom Resource Usage Monitor

```python
def monitor_resource_usage(resource_usage):
    overused_resources = []

    # Check each resource usage
    for resource, hours in resource_usage.items():
        if hours > 8:
            overused_resources.append(resource)

    # Display results
    if overused_resources:
        print("Overused Resources:", ", ".join(overused_resources))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")
```

```python
usage_data = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

monitor_resource_usage(usage_data)
```

Problem Statement 6: Online Event Registration Capacity Controller

```python
def manage_event_registration(capacity, total_registrations):
    confirmed = 0
    waitlisted = 0

    # Check if registrations exceed capacity
    if total_registrations > capacity:
        confirmed = capacity
        waitlisted = total_registrations - capacity
        status = "Closed"
    else:
        confirmed = total_registrations
        waitlisted = 0
        status = "Open"

    # Display results
    print(f"Confirmed Registrations: {confirmed}")
    print(f"Waitlisted Users: {waitlisted}")
    print(f"Registration Status: {status}")
```

```python
manage_event_registration(100, 105)
```
